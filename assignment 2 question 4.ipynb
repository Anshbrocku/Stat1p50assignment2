```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('students.csv')

data2 = data[['Programming skill','Theoretical knowledge', 'Class']]

data2 = data2.groupby('Class').mean().reset_index()

print(data2)
plt.bar(x=data2.loc[:,'Class'],height=data2.loc[:,'Programming skill'],label='Programming skill')
plt.legend()
plt.show()
plt.bar(x=data2.loc[:,'Class'],height=data2.loc[:,'Theoretical knowledge'],label='Theoretical knowledge')
plt.legend()
plt.show()

```
