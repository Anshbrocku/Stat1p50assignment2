```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('crime.csv')
x= data[['PctPopUnderPov']]
y=data[['ViolentCrimesPerPop']]
plt.scatter(x,y,color='red')
import numpy as np
crime_list1 = np.array(data['PctPopUnderPov']).reshape(-1,1)
crime_list = np.array(data['ViolentCrimesPerPop']).reshape(-1,1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
crime_list = scaler.fit_transform(crime_list)
crime_list1 = scaler.fit_transform(crime_list1)

plt.scatter(crime_list1,crime_list,color='blue')
```
