```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('students2.csv')

data2 = data[['Programming skill','Theoretical knowledge', 'Class']]

data2 = data2.groupby('Class').mean().reset_index()

print(data2)
plt.plot(data2.loc[:,'Class'],data2.loc[:,'Programming skill'],'o-',label='Programming skill',color='red')
plt.legend()
plt.show()
plt.plot(data2.loc[:,'Class'],data2.loc[:,'Theoretical knowledge'],'o-',label='Theoretical knowledge',color='orange')
plt.legend()
plt.show()
#the program will work fine for more number of the classes as well
```
